concat은 포개지지 않고 붙인다면 join, merge는 겹치는 컬럼에 대해서 값을 기준으로 데이터를 합침

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2']},
                   index=['K0', 'K1', 'K2'])
```

```python
right = pd.DataFrame({'C': ['C0', 'C1', 'C2'],
                    'D': ['D0', 'D1', 'D2']},
                   index=['K0', 'K2', 'K3'])
```

```python
left
right
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                   'key': ['K0', 'K1', 'K0', 'K1']})

right = pd.DataFrame(
    {
        'C': ['C0', 'C1'],
        'D': ['D0', 'D1'],    
    },
    index=['K0', 'K1']
)
```

```python
left
right
```

```python
# right를 key 컬럼에 맞춤
left.join(right, on='key')
```

```python
left.join(right, on='key').set_index('key')
```

```python
left.set_index('key').join(right)
```
