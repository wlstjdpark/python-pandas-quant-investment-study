lec1_3에서 median 시가총액 연결하기

```python
a_df = pd.read_csv('data/my_data/Small_and_Big.csv', index_col=[0])
a_df.head()
```

```python
median_df = a_df.groupby(['date']).agg({'시가총액 (보통)(평균)(원)': 'median'})
median_df.columns = ['시가총액_median']
median_df.head()
```

```python
a_df.set_index('date').join(median_df, on='date').head()
```

```python
joined_df = a_df.join(median_df, on='date')
joined_df.head()
```
