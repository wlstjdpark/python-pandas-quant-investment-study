```python
import pandas as pd
import numpy as np
```

```python
s1 = np.arange(1, 6, 1)
s2 = np.arange(6, 11, 1)
s1
s2
```

```python
pd.DataFrame({'c1': s1, 'c2': s2})
```

```python
pd.DataFrame([
    [10,11],
    [10,12]
])
pd.DataFrame(
    np.array(
        [
            [10, 11],
            [10, 12]
        ]
    )
)
```

```python
pd.DataFrame([
    np.arange(1, 100),
    np.arange(100, 200)
])
```

```python
pd.DataFrame(
    {
        'c1': pd.Series(np.arange(1, 100)),
        'c2': pd.Series(np.arange(100, 200))
    }
)
```

```python
s1 = pd.Series(np.arange(1,6,1))
s2 = pd.Series(np.arange(6,11,1))
pd.DataFrame({'c1': s1, 'c2': s2})
```

```python
s1 = pd.Series(np.arange(1,6,1), index=['a','b','c','d','e'])
s2 = pd.Series(np.arange(6,11,1), index=['b','c','d','f','g'])
pd.DataFrame({'c1': s1, 'c2': s2})
```
